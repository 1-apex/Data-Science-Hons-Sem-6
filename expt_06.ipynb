```python
import nltk
from nltk.corpus import stopwords
import os
import numpy as np
import re
from collections import Counter

os.chdir("C:\\Users\\prath\\Documents\\Python\\Data Science HONS SEM 6\\Expt 06\\train-mails")

curr_dic = os.getcwd()
```

```python
# creating a dictionary of words from training data
def make_dictionary():
    emails = [os.path.join(curr_dic,f) for f in os.listdir(curr_dic)]
    all_words = []

    stop_words = set(stopwords.words('english'))

    for mail in emails:
        with open(mail) as m:
            for i, line in enumerate(m):
                # Body of the email start at line 3
                if i == 2:                
                    # remove punctuation and symbols
                    line = re.sub(r'[^a-zA-Z0-9\s]', '', line)

                    # remove digits
                    line = re.sub(r'\d', '', line)

                    words = line.split()

                    for word in words:
                        if word in stop_words or len(word) <= 1:
                            words.remove(word)

                    all_words += words

    # we have the count of all the words
    dictionary = Counter(all_words)
    return dictionary
```
